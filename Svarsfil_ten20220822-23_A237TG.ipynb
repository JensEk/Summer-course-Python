```python
#Funktionsdefinition

# Tar två int som argument och beräknar volym enligt angiven formula med avrundning 1 decimal
def volymPyramid(B, h):
    
    volym = round((B*h/3), 1)
    return volym

#Huvudprogram

print('-- Beräkna volymen på en pyramid --')
basArea = int(input('Basarea: '))
hojd = int(input('Höjd: '))
beraknadVolym = volymPyramid(basArea, hojd)
print(f'- Volymen är {beraknadVolym}')
```

```python
import random

# Huvudprogram

slumpLista = []
minVal = 11    # Eftersom rangen är 1-10 så är alla tal mindre än 11
maxVal = 0    # Eftersom rangens minsta tal är större än 0
totalVal = 0    # Håller totalen av slumpade talen

# Appendar en random int mellan 1 och 10 fem gånger
for i in range (5):
    slumpLista.append(random.randint(1,10))

print(f'Slumplistan blev: \n{slumpLista}')

# Loopar igenom hela slumpade listan och uppdaterar de globala variablerna minVal och maxVal beroende på villkoren
# Inkrementerar totalVal med varje värde i lista för att kunna beräkna medelvärdet senare
for i in range(len(slumpLista)):
    if(slumpLista[i] < minVal):
        minVal = slumpLista[i]
    if(slumpLista[i] > maxVal):
        maxVal = slumpLista[i]
    totalVal += slumpLista[i]

print(f'Minsta talet: {minVal}')
print(f'Största talet: {maxVal}')
print(f'Medelvärde: {totalVal/len(slumpLista)}')
    
```

```python
import csv

#Funktionsdefinition:

# Tar filnamn som argument och returnerar en array med datan inläst med separation vid ';' tecknet
def read_file(file_name):
    dataRead = []
    
    with open(file_name, 'r', encoding='UTF-8') as file:
            csv_reader = csv.reader(file, delimiter=';')
            
            for rad in csv_reader:
                dataRead.append(rad)
    return dataRead

# Huvudprogram
cars_data = read_file('cars.csv')
print(cars_data[0:5], '\n')    # Skriver ut de första 5 raderna i listan som angivet
```

```python
import matplotlib.pyplot as plt

# Skapar två dictionarys för vardera objekt
am = {}
gear = {}

# Går över alla objekt i inlästa datan och lägger till key för motsvarande am(0 eller 1) och inkrementerar sedan dess value med 1 för antalet bilar.
# Gör samma för motsvarande gear(3,4,5) och inkrementerar dess value med 1 för antalet bilar.
for i in range(1, len(cars_data), 1): 
    if not cars_data[i][6] in am:
        am[cars_data[i][6]] = 0
    am[cars_data[i][6]] += 1
    
    if not cars_data[i][7] in gear:
        gear[cars_data[i][7]] = 0
    gear[cars_data[i][7]] += 1

    
print('-'*40)

print('Automatic/Manuel (am) \t Count of cars')
for typ,antal in am.items():   # Returnerar varje key/value par i am och skriver ut enligt anvisning
    print(f'{typ:<30} {antal: >2}')

print('-'*40)
print('gear \t \t \t Count of cars')
print('-'*40)

for typ,antal in gear.items():    # Returnerar varje key/value par i gear och skriver ut enligt anvisning
    print(f'{typ:<30} {antal:>2}')


# Returnerar både dictionarys key/value par som listor att använda i plot
amKeys = list(am.keys())
antalFordonAM = list(am.values())
gearKeys = list(gear.keys())
antalFordonGear = list(gear.values())

plt.figure(figsize=(7,8))

# subplot med 2 rader och 1 kolumn
# rad 1 som plottar am och dess antal
plt.subplot(2, 1, 1)
plt.xlabel('am')
plt.ylabel('Count of Cars')
plt.bar(amKeys, antalFordonAM, color='blue', width=0.5)
plt.grid()

# rad 2 som plottar gear och dess antal
plt.subplot(2, 1, 2)
plt.xlabel('gear')
plt.ylabel('Count of Cars')
plt.bar(gearKeys, antalFordonGear, color='green', width=0.8)
plt.grid()
```

```python
# Huvudprogram

# Skapar dictonary där key blir 'bilnamnet' och value en array med am och gear värdet i.
cars = {}

# listor som håller vardera värde för motsvarande bil, etc index 0 är bil nummer 1 ("Cadillac Fleetwood")
hp = []   
mpg = []
cyl = []

for i in range(1, len(cars_data), 1):
    if not cars_data[i][0] in cars:
        cars[cars_data[i][0]] = [cars_data[i][6], cars_data[i][7]]
    mpg.append(float(cars_data[i][1].replace(',','.')))
    cyl.append(cars_data[i][2])
    hp.append(cars_data[i][4])
    

# Diagram 1

# Returnerar namnet, am och gear som separata listor att använda till plott
carName = list(cars.keys())
carValues = list(cars.values())

carAm = []
carGear = []

for i in range(len(carValues)):
    carAm.append(carValues[i][0])
    carGear.append(carValues[i][1])

# plot am/gear mot motsvarande bilmärke som uppgift angivelsen
plt.figure(figsize=(10,5))
plt.title('Line plot for am and gear for different cars')

plt.plot(carName, carAm, color="blue", linestyle = '-')    # Första med am
plt.plot(carName, carGear, color="orange", linestyle = '-')    # Andra med gear
plt.xticks(rotation = 90)    # rotera x namnen 90 grader 
plt.legend(["am", "gear"], loc="upper left")    # Lägg till rubriker för datan
plt.grid()
plt.show()


# Diagram 2

# Plottar hp mot respektive mpg med tillhörande bilnamn vid markeringen

plt.figure(figsize=(15,10))
plt.scatter(mpg, hp, c ="red")
plt.title('Hp vs mpg for different cars')
plt.xlabel('mpg')
plt.ylabel('hp')

for i, name in enumerate(carName):    # Går över all bilnamn och appointar dem till respektive datapunkt i mpg och hp
    plt.annotate(name, (mpg[i], hp[i]))

plt.xticks(mpg)    # Lägger till alla värden för mpg i xaxeln
plt.show()



# Diagram 3

# Plottar hp mot antalet cylindrar för varje bilnamn

plt.figure(figsize=(12,12))
plt.bar(cyl, hp, width=0.6 ,color="steelblue")
plt.title('hp and cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('hp')
plt.show()
```
