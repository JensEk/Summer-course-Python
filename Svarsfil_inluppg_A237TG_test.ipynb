```python
# Funktion för att läsa in en angiven csv-fil

def read_file(file_name):
    import csv
    file_name_list=[]                              #skapar en tom lista...
    with open(file_name) as file:
        csv_reader=csv.reader(file, delimiter=';')   #läser in filen och sorterar listan som separeras av ';'   
        
        for row in csv_reader:
            file_name_list.append(row)             # listan fylls med raderna från csv filen
            
    return file_name_list                          #funktionen returnerar listan


# Huvudprogram

kameradata=read_file('kameraData.csv')
print(kameradata[0:3], '\n')

platsdata=read_file('platsData.csv')
print(platsdata[:3], '\n')
```

```python
#Funktion

def antal_bilar(kamera_data):
    kamera_data=['40', '50', '60', '70', '80', '90']     #skapar en lista för att jämföra med kameraData.csv
    x1=0 
    x2=0
    x3=0                                     # skapar en variabel för respektive hastighet
    x4=0
    x5=0
    x6=0
    for col in kameradata:                  #skapar en for-loop som räknar förekomsten för varje hastighet
        
        for e in col[:2]:                    #inkludera kolumnera 0-1
            if (e==kamera_data[0]):
                x1+=1
                
            elif(e==kamera_data[1]):
                x2+=1
                
            elif(e==kamera_data[2]):
                x3+=1
                
            elif(e==kamera_data[3]):
                x4+=1
            
            elif(e==kamera_data[4]):
                x5+=1
            
            elif (e==kamera_data[5]):
                x6+=1
                
    print(f' Det finns {x1} mätningar där gällande hastighet är {kamera_data[0]} km/h')        #utskrift
    print(f' Det finns {x2} mätningar där gällande hastighet är {kamera_data[1]} km/h')
    print(f' Det finns {x3} mätningar där gällande hastighet är {kamera_data[2]} km/h')
    print(f' Det finns {x4} mätningar där gällande hastighet är {kamera_data[3]} km/h')
    print(f' Det finns {x5} mätningar där gällande hastighet är {kamera_data[4]} km/h')
    print(f' Det finns {x6} mätningar där gällande hastighet är {kamera_data[5]} km/h')
    print('---------------------------------------------------------------')
    print(f'Totalt paserade {x1+x2+x3+x4+x5+x6} bilar')
    
    import matplotlib.pyplot as plt
    
    hastigheter=[x1, x2, x3, x4, x5, x6]      #skapar en lista med hastigheterna för plotningen
    
    plt.bar(kamera_data, hastigheter, width=0.1)                 # justerar storlekar, titlar, osv.....
    plt.grid()
    
    plt.title('Antal fordon för varje gällande hastighet')
    plt.xlabel('Gällande hastighet')
    plt.ylabel('Antal passerande fordon')
    plt.show()
    
    

#Huvudprogram


antal_bilar(kameradata)
```

```python
def antal_kameror(plats_data):
    

    kommuner=[]                                     #skapar en tom lista för alla kommuner
    for col in plats_data:
        for e in col[3:4]:                          # enbart inkludera 4:e kolumnen (kommuner)
            kommuner.append(e)                      # lägger in kommuner på en ny lista

    kommun=[]                                       #skapar en ny lista för varje enskild kommun

    for i in set(kommuner):
        kommun.append(i)
    kommun.remove('Kommun')                         # tar bort strängen "kommun" ur listan
    kommun.sort()

    antalkameror=[]                                 # ny lista för antalet kameror

    index=0
    z=0                                             #räknare

    print(f' Kommun \t Antal kameror')
    print(f'-------------------------------')


    while index<len(kommun):                        #skapar en while-loop som +antalet kameror för varje kommun
        for i in kommuner:
            if i==kommun[index]:
                z+=1
        antalkameror.append(z)                      # lägger till antalet kameror i listan 
    
        print(f'{kommun[index]:12} {antalkameror[index]:10}')
        index+=1                                    
        z=0                                         # nolställer räknaren inför nästa loop
        continue


    print('------------------------')
    totalt_antal_kameror=len(kommuner)
    print(f'Det finns totalt {totalt_antal_kameror} kameror')


#Huvudprogram

antal_kameror(platsdata)
```

```python
#Uppgift 4a)

def hastighetplus(kamera_data):

    print(
    'Söker ut alla poster där den procentuella hastighetsöverträdelsen'
        ' överskrider\ngällande hastighet med mer än vald %-sats...')
    
    
    procentover=float(input('Med hur många % skall hastigheten överskrida hastighetsbegränsningen'
                        'för att visas?  '))          #Gör användarens input till float tal
    print('\n')


    gh=[]                                            #skapar två nya listor för respektive typ av mätning
    fh=[]

    for col in kamera_data:
        for e in col[1:2]:
            gh.append(e)
    
        for e in col[2:3]:
            fh.append(e)
        
    del(gh[0])                                            #tar bort rubrikerna ur de nya listorna
    del(fh[0])

    for i in range(0, len(gh)):                          #gör innehållet i listorna till typen int
        gh[i]=int(gh[i])

    for i in range(0, len(fh)):
        fh[i]=int(fh[i])
    
    överträdelser=[]                                #skapar en ny lista som visar på vilka index överträdelser sker
    räknare=0

    for item in fh:
        if item>(gh[räknare]*((procentover/100)+1)):
            överträdelser.append(räknare+1)          #adderar 1 för att kompensera för rubriken som togs bort...
        räknare+=1                                          #... för att få rätt index i kameradata
    
    
    plushastighet=[]

    for i in överträdelser:
        plushastighet.append(kameradata[i])
    
    print(f'Det var {len(plushastighet)} överträdelser som var mer än {procentover:0.0f}% över gällande hastighet.', '\n')

    print(f'{"Tid":<10}{"MätPlatsID":<15}{"Gällande hastighet":<20} {"Hastighet":<15} {"Datum"}')
    print(f'--------------------------------------------------------------------------')

    for e in plushastighet:                      #lägger tiden först i varje element
        e.reverse()

    plushastighet.sort()                         #sorterar listans efter elementet "tid"

    for e in plushastighet:                      #roterar tillbaks listan
        e.reverse()

    for e in plushastighet:
        print(f'{e[4]:<10} {e[0]:<20} {e[1]:<15} {e[2]:<10} {e[3]}')
        
    return procentover, plushastighet                  #returnerar procentover och plushastighet
                                                        #för kommande funktioner

#----------------------------------------------------------------------------------------------------------------    
    
#Uppgift 4b)

def hastighetplusdiagram(plus_hastighet, procent_over):
    
    
    import matplotlib.pyplot as plt


    tid=[]
    faktisk=[]                             #listor för faktiska hastigheter, gällande hastighet och tid
    gällande=[]


    for e in plus_hastighet:               #for-loop som fyller listorna med specifikt element från plushastighet
        faktisk.append(e[1])
        tid.append(e[4])
        gällande.append(e[2])


    for i in range(0, len(faktisk)):             #gör hastighetslistorna till integer
        faktisk[i]=int(faktisk[i])
    
    for i in range(0, len(gällande)):
        gällande[i]=int(gällande[i])
    
    fig=plt.figure()
    ax=fig.add_subplot(111)                            #gör subplottning möjlig 

    ax.barh(tid, gällande, height=0.4, color= 'g', align='center')            #justerar storlek, färg, osv...
    ax.barh(tid, faktisk, height=0.8, color='k', align='center')
    title_font = {'size':'8'}
    plt.xlim(0,250)
    plt.grid()
    
    plt.ylabel('Tid')                                  #justerar titlarna
    plt.xlabel('Hastighet')
    plt.title(f'Tidpunkter där gällande hastighet överskridits med mer än {procent_over:0.0f}%', **title_font )

    plt.show()    

#----------------------------------------------------------------------------------------------------------------

#Uppgift 4c)

def hastighetkommun(plus_hastighet, plats_data):

    mätid=[]                               #skapar en lista för mätningarnas ID

    for e in plus_hastighet:              #denna lista fylls med de ID där hastighetsöverträdelser inträffat       
        mätid.append(e[0])  

    kommun_vägnummer=[]                      #skapar en lista för element från platsdata
    

    for e in mätid:                       #hämtar de element från platsdata som matchar 
        for i in plats_data:                # mätningarnas ID i plushastighet
            if e==i[0]:
                kommun_vägnummer.append(i)

                                               #utskrift som är formaterad enligt uppgiftsbeskrivning
                    
    print(f'Hastighetsöverträdelserna skedde vid följande {len(mätid)} platser', '\n')

    print(f'{"Tid":<10}{"MätPlatsID":<15}{"Kommun":<10} {"Vägnummer":<15} {"Hastighet"}')
    print('----------------------------------------------------------------')

    for e,i in zip(plushastighet, kommun_vägnummer):
        print(f'{e[4]:<10} {e[0]:<13} {i[3]:<13} {i[2]:<15} {e[2]}')
    
    
#Huvudprogram

#a)
procentover, plushastighet= hastighetplus(kameradata)

#b)
hastighetplusdiagram(plushastighet, procentover)

#c)
hastighetkommun(plushastighet, platsdata)
```

```python
def bilar_timme(kamera_data):

    t=[]                                       #lista som fylls med alla tider från kolumn 4 i kameradata

    for col in kamera_data:
        t.append(col[4])
   
    del(t[0])                                 #tar bort elementet "tid" och sorterar listan 
    t.sort()                                    
    
    from datetime import datetime
    
    tid=[]                                   #skapar en ny lista för datetime 

    for e in t:
        tid.append(datetime.strptime(e, '%H:%M:%S').hour)    #fyller och anger formatet H,M,S för listan tid
            
    antal_fordon=[]          #ny lista för antal förekomster vid varje timslag

    x=7
    for e in tid:            #räknare som börjar på 7 och räknar förekomsten av varje tid och lägger det i lista
        if e==x:
            antal_fordon.append(tid.count(x))
            x+=1

    import matplotlib.pyplot as plt

                                                  #skapar x-kordinater för rätt format i x-axeln 
    x_koord=['07:00', '08:00', '09:00', '10:00', '11:00', '12:00', '13:00', '14:00', '15:00', '16:00', '17:00']
    
    plt.grid()                                    #justerar formatet för plotningen
    plt.ylim(0,180)
    plt.title('Antal fordon per timma mellan kl. 07.00 och 17.00')
    plt.ylabel('Antal fordon')
    plt.xlabel('Tid')
    
    plt.plot(x_koord, antal_fordon)
    plt.show()
    
    
    
#Huvudprogram

bilar_timme(kameradata)
```
